# plant 파생변수 생성

## 이슬점 추가하기

```python
import sys
import numpy as np
 
# approximation valid for
# 0 degC < T < 60 degC
# 1% < RH < 100%
# 0 degC < Td < 50 degC 
 
# constants
a = 17.271
b = 237.7 # degC
def dewpoint_approximation(T,RH):
 
    Td = (b * gamma(T,RH)) / (a - gamma(T,RH))
 
    return Td
 
def gamma(T,RH):
 
    g = (a * T / (b + T)) + np.log(RH/100.0)
 
    return g
def inAppendDewpoint(data):
    data["DEW_POINT_IN"] = dewpoint_approximation(data["TEM_IN"], data["HUM_IN"])
    return data
def OutAppendDewpoint(data):
    data["DEW_POINT_OUT"] = dewpoint_approximation(data["TEM_OUT_LOC1"], data["HUM_OUT_LOC1"])
    return data
def CoilAppendDewpoint(data):
    data["DEW_POINT_COIL"] = dewpoint_approximation(data["TEM_COIL"], data["HUM_IN"])
    return data
```

```python
plant = inAppendDewpoint(plant)
plant = OutAppendDewpoint(plant)
plant = CoilAppendDewpoint(plant)
```

```python
plant
```

## 포화 수증기압 , 수증기압

```python
import math
def saturationVaporPressure(t):
    '''포화 수증기압 구하는 공식'''
    t = float(t)
    return round(6.11* math.exp((17.27*t)/(237.3+t)),2)
```

```python
plant["STR_WVP"]= plant["TEM_IN"].map(lambda x :saturationVaporPressure(x))
plant
```

### 수증기 압

```python
plant["WVP"] = plant["TEM_IN"]*plant["STR_WVP"] / 100
plant
```

## 내부온도 - 외부 온도 

```python
plant["IN_OUT_TEM_D"] = plant["TEM_IN"] - plant["TEM_OUT_LOC1"]
plant
```

```python
plant["IN_COIL_TEM_D"] = plant["TEM_IN"] - plant["TEM_COIL"]
plant
```

## 내부 습도 - 외부 습도

```python
plant["IN_OUT_HUM_D"] = plant["HUM_IN"] - plant["HUM_OUT_LOC1"]
plant.head()
```

## 데이터 시계열 특성 추가
### 데이터 형식 바꿔주기

```python
def changeToDatatime(data):
    data["MEA_DDHR"] = pd.to_datetime(data['MEA_DDHR'])
    data['X24H_TMA'] = pd.to_datetime(data['X24H_TMA'])
    data['X48H_TMA'] = pd.to_datetime(data['X48H_TMA'])
    return data
```

```python
plant = changeToDatatime(plant)
plant.dtypes
```

### 월, 달, 시 추가해주기

```python
def appendMonthANdWeek(data):
    data['MONTH'] = data['MEA_DDHR'].dt.month
    data['WEEK'] = data['MEA_DDHR'].dt.week
    data['HOUR'] = data['MEA_DDHR'].dt.hour
    return data
```

```python
plant = appendMonthANdWeek(plant)
plant = plant.drop(columns=["COND_LOC"],)
plant
```

## 기상변수 추가하기

```python
def changeWeatherFeature(data):
    data.columns = data.columns.str.upper()
    data['TMA'] = pd.to_datetime(data['TMA'])
    data = data.sort_values(by=['TMA'])
    data.rename({'TMA':'MEA_DDHR'}, axis='columns', inplace=True)
    return data
```

```python
weather = changeWeatherFeature(weather)
weather
```

```python
weather.to_csv("data/32.finaldata_weather/aws_merge_selected.csv", index=False)
```

```python
def etlPlantWeather(plant, weather):
    plant_weather = plant.merge(weather, how='left', left_on='MEA_DDHR', right_on='MEA_DDHR')
    plant_weather = plant_weather.interpolate(method="linear")
    return plant_weather
```

```python
plant_weather = etlPlantWeather(plant, weather)
plant_weather
```

## 해상변수 추가하기

```python
sea = changeWeatherFeature(sea)
sea
```

```python
sea.to_csv("data/32.finaldata_weather/sea_merge_selected.csv", index=False)
```

```python
plant_weather_sea = etlPlantWeather(plant_weather, sea)
plant_weather_sea
```
