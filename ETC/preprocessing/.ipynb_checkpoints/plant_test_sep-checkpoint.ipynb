## plant2_test

### loc1, 2, 3 나누기

```python
import pandas as pd
plant2_test = pd.read_csv('../data/plant2_test.csv')
```

```python
del plant2_test['plant']
```

```python
loc1 = plant2_test[plant2_test['loc'] == 1].copy()
loc1.head()
```

```python
del loc1['loc']
```

```python
loc1.columns
```

```python
loc1.columns = ['mea_ddhr', 'tem_in_loc1', 'hum_in_loc1', 'tem_coil_loc1', 'tem_out_loc1',
       'hum_out_loc1', 'x24h_tma', 'x24h_cond_loc', 'x48h_tma',
       'x48h_cond_loc']
```

```python
loc1.head()
```

```python
loc2 = plant2_test[plant2_test['loc'] == 2].copy()
loc2.head()
```

```python
del loc2['loc']
```

```python
loc2.columns
```

```python
loc2.columns = ['mea_ddhr', 'tem_in_loc2', 'hum_in_loc2', 'tem_coil_loc2', 'tem_out_loc1',
       'hum_out_loc1', 'x24h_tma', 'x24h_cond_loc', 'x48h_tma',
       'x48h_cond_loc']
```

```python
loc2.head()
```

```python
loc3 = plant2_test[plant2_test['loc'] == 3].copy()
loc3.head()
```

```python
del loc3['loc']
```

```python
loc3.columns
```

```python
loc3.columns = ['mea_ddhr', 'tem_in_loc3', 'hum_in_loc3', 'tem_coil_loc3', 'tem_out_loc1',
       'hum_out_loc1', 'x24h_tma', 'x24h_cond_loc', 'x48h_tma',
       'x48h_cond_loc']
```

```python
loc3.head()
```

### loc1, 2, 3 merge

```python
loc1.shape
```

```python
loc2.shape
```

```python
loc3.shape
```

```python
plant2_test = loc1.merge(loc2, on='mea_ddhr', how='outer')
```

```python
plant2_test = plant2_test.merge(loc3, on='mea_ddhr', how='outer')
```

```python
plant2_test.head()
```

```python
plant2_test.shape
```

### columns 정리

```python
plant2_test.columns
```

```python
plant2_test = plant2_test[['mea_ddhr', 'tem_in_loc1', 'hum_in_loc1', 'tem_in_loc2', 'hum_in_loc2', 'tem_in_loc3',
       'hum_in_loc3','tem_out_loc1_x', 'hum_out_loc1_x', 'tem_out_loc1_y', 'hum_out_loc1_y', 'tem_out_loc1', 'hum_out_loc1', 
        'tem_coil_loc1', 'tem_coil_loc2', 'tem_coil_loc3', 'x24h_tma_x', 'x24h_cond_loc_x',
       'x48h_tma_x', 'x48h_cond_loc_x', 
        'x24h_tma_y','x24h_cond_loc_y', 'x48h_tma_y', 'x48h_cond_loc_y',   
       'x24h_tma', 'x24h_cond_loc', 'x48h_tma', 'x48h_cond_loc']]
```

```python
plant2_test.head()
```

```python
min(plant2_test['mea_ddhr'])
```

```python
plant2_test = plant2_test.sort_values(by='mea_ddhr')
```

```python
plant2_test.head()
```

```python
plant2_test = plant2_test.reset_index()
```

```python
plant2_test.head()
```

```python
del plant2_test['index']
```

```python
pd.set_option('display.max_columns', 30)
```

```python
plant2_test.head()
```

```python
plant2_test.dtypes
```

```python
fillna_0_cols = ['tem_out_loc1_x', 'hum_out_loc1_x', 'tem_out_loc1_y', 'hum_out_loc1_y', 'tem_out_loc1', 'hum_out_loc1']
```

```python
plant2_test[fillna_0_cols] = plant2_test[fillna_0_cols].fillna(0)
```

```python
plant2_test.head()
```

```python
plant2_test['tem_out_loc1'] = plant2_test['tem_out_loc1_x'] + plant2_test['tem_out_loc1_y'] + plant2_test['tem_out_loc1']
```

```python
plant2_test.head()
```

```python
del plant2_test['tem_out_loc1_x']
del plant2_test['tem_out_loc1_y']
```

```python
plant2_test.head()
```

```python
plant2_test['hum_out_loc1'] = plant2_test['hum_out_loc1_x'] + plant2_test['hum_out_loc1_y'] + plant2_test['hum_out_loc1']
```

```python
del plant2_test['hum_out_loc1_x']
del plant2_test['hum_out_loc1_y']
```

```python
plant2_test.head()
```

```python
# plant2_test['x24h_tma'] = plant2_test['x24h_tma_x'] + plant2_test['x24h_tma_y'] + plant2_test['x24h_tma']
# del plant2_test['x24h_tma_x']
# del plant2_test['x24h_tma_y']
```

```python
date = plant2_test.copy()
```

```python
date_na_merge = date[['x24h_tma_x', 'x24h_tma_y', 'x24h_tma']]
```

```python
# date['x24h_tma'] = date['x24h_tma_x'] + date['x24h_tma_y'] + date['x24h_tma']
```

```python
date.head()
```

```python
# first_valid_index
date_na_merge.first_valid_index
```

```python
date_na_merge.head()
```

```python
def get_first(row) :
    return row[row.first_valid_index()]
```

```python
date['x24h_tma'] = date_na_merge.apply(lambda row : row[row.first_valid_index()], axis=1)
```

```python
date_na_merge2 = date[['x48h_tma_x', 'x48h_tma_y', 'x48h_tma']]
```

```python
date_na_merge2.head()
```

```python
date['x48h_tma'] = date_na_merge2.apply(lambda row : row[row.first_valid_index()], axis=1)
```

```python
date.head()
```

```python
date.head()
```

```python
date.shape
```

```python
date.isnull().sum()
```

```python
date.columns
```

```python
date = date.drop(['x24h_tma_x',
       'x24h_cond_loc_x', 'x48h_tma_x', 'x48h_cond_loc_x', 'x24h_tma_y',
       'x24h_cond_loc_y', 'x48h_tma_y', 'x48h_cond_loc_y'], axis=1)
```

```python
date.columns
```

```python
date.head()
```

```python
plant2_test = date.copy()
```

```python
plant2_test.to_csv('../data/plant2_test_reshape.csv', index=False, encoding='utf-8')
```
